```python
#now lets do some EDA on the preprocessed words
#will write a basic preprocess function using a stemmer and then count-vectorize the resulting corpus
def preprocess_stem(text):
    #Tokenize text
    tokens = word_tokenize(text.lower())
    
    #define stop words and punctuation
    stop_words = stopwords.words('english')
    punc = list(string.punctuation)
    
    #Stem the tokens
    stemmer = PorterStemmer()
    stemmed_tokens = [stemmer.stem(token) for token in tokens if token not in stop_words and token not in punc]
    
    return ' '.join(stemmed_tokens)    
    
```

```python
#lets get the stemmed titles for fake and real articles
titles_stemmed = articles['title'].apply(preprocess_stem)
```

```python
#now lets count_vectorize the stemmed titles
vectorizer = CountVectorizer()
vectorized_df = vectorizer.fit_transform(titles_stemmed)
```

```python
#lets find the 50 most common tokens in all articles and create a bar plot of their values

#sums the counts of each token across all titles
total_token_counts = np.array(vectorized_df.sum(axis=0))[0] / len(articles)

#finds the indices of the top 50 tokens
top_ind = np.argpartition(total_token_counts, -25)[-25:]

#gets the tokens corresponding to these maximum counts and the counts
top_tokens = vectorizer.get_feature_names_out()[top_ind]
top_tokens_counts = total_token_counts[top_ind]
```

```python
#now lets make a barplot of these values in sorted order
fig, ax = plt.subplots(figsize=(12, 6))

_ = sns.barplot(x=top_tokens, y=top_tokens_counts, palette = 'cool',
               order = [token for _, token in sorted(zip(top_tokens_counts, top_tokens), reverse=True)])
ax.bar_label(ax.containers[0], fmt='%.2f')
_ = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
_ = ax.set(xlabel = 'Token', ylabel = 'Proportion', title = 'Proportions of 25 Most Frequent Tokens in All Articles')
```

```python
#now lets start the analysis 

#first i will write a preprocess function that lemmatize's the titles instead of stemming

def preprocess_lemmatize(text):
    #Tokenize text
    tokens = word_tokenize(text.lower())
    
    #define stop words and punctuation
    stop_words = stopwords.words('english')
    punc = list(string.punctuation)
    
    #Lemmatize the tokens
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stop_words and token not in punc]
    
    return ' '.join(lemmatized_tokens)          
```

```python
#next i will define some dictionaries that should help in running the models

param_preprocessing = {
    'stem': preprocess_stem,
    'lem': preprocess_lemmatize
}

param_vectorizers = {
    'count_vectorizer': CountVectorizer(),
    'tfidf_vectorizer': TfidfVectorizer()
}
```

```python
X_train, X_test, y_train, y_test = train_test_split(articles['title'], articles['fake'], test_size=0.25,
                                                   random_state=42) 
```

```python
#lets start with a Naive Bayes classifier
nb_names = []
nb_grids = []
nb_best_params = []
nb_scores = []
nb_times = []

for prep_name, prep_function in param_preprocessing.items():
    
    print(f'Applying {prep_name}...')
    
    tokenized_train = X_train.apply(prep_function)
    tokenized_test = X_test.apply(prep_function)
    
    for vectorizer_name, vectorizer in param_vectorizers.items():
        print(f'Running {vectorizer_name}...')
        
        #create pipeline
        pipeline = Pipeline([
            ('vectorizer', vectorizer),
            ('model', MultinomialNB())
        ])
        
        #define the params for GridSearch to iterate over
        param_grid = {
            'vectorizer__max_features': [1000, 2000, 5000, None],
            'model__alpha': [.001, .01, .1, 1]
        }
        
        #create and fit the grid and also time its fitting time
        grid = GridSearchCV(pipeline, param_grid)
        start_time = time.time()
        grid.fit(tokenized_train, y_train)
        end_time = time.time()
        
        #add the name of this run, the grid, the best params, and best score the lists to keep track
        nb_names.append(prep_name + '_' + vectorizer_name + '_' + 'NativeBayes')
        nb_grids.append(grid)
        nb_best_params.append(grid.best_params_)
        nb_scores.append(grid.best_estimator_.score(tokenized_test, y_test))
        nb_times.append(end_time-start_time)
```

```python
nb_results = pd.DataFrame({'model': nb_names,
                          'best_params': nb_best_params,
                          'train_times (s)': nb_times,
                          'best_score': nb_scores}).set_index('model').sort_values('best_score', ascending=False)

nb_results
```

```python
lgr_results = pd.DataFrame({'model': lgr_names,
                          'best_params': lgr_best_params,
                          'train_times (s)': lgr_times,
                          'best_score': lgr_scores}).set_index('model').sort_values('best_score', ascending=False)

lgr_results
```

```python
##now lets fit a SVM Model
svm_names = []
svm_grids = []
svm_best_params = []
svm_scores = []
svm_times = []

for prep_name, prep_function in param_preprocessing.items():
    
    print(f'Applying {prep_name}...')
    
    tokenized_train = X_train.apply(prep_function)
    tokenized_test = X_test.apply(prep_function)
    
    for vectorizer_name, vectorizer in param_vectorizers.items():
        print(f'Running {vectorizer_name}...')
        
        #create pipeline
        pipeline = Pipeline([
            ('vectorizer', vectorizer),
            ('model', SVC(max_iter=10000))
        ])
        
        #define the params for GridSearch to iterate over
        param_grid = {
            'vectorizer__max_features': [2000, None],
            'model__kernel': ['rbf'],
            'model__gamma': ['scale'],
            'model__C': [1, 10]
        }
        
        #create and fit the grid and also time its fitting time
        grid = GridSearchCV(pipeline, param_grid)
        start_time = time.time()
        grid.fit(tokenized_train, y_train)
        end_time = time.time()
        
        #add the name of this run, the grid, the best params, and best score the lists to keep track
        svm_names.append(prep_name + '_' + vectorizer_name + '_' + 'SVM')
        svm_grids.append(grid)
        svm_best_params.append(grid.best_params_)
        svm_scores.append(grid.best_estimator_.score(tokenized_test, y_test))
        svm_times.append(end_time-start_time)
```

```python
#now lets try a Random Forest model
rf_names = []
rf_grids = []
rf_best_params = []
rf_scores = []
rf_times = []

#in order to save time on the RF model, we will just use a stemmer and a tdidf vectorizer        
print("Applying stem...")
tokenized_train = X_train.apply(preprocess_stem)
tokenized_test = X_test.apply(preprocess_stem)

#create pipeline
pipeline = Pipeline([
    ('vectorizer', TfidfVectorizer()),
    ('model', RandomForestClassifier())
])

#define the params for GridSearch to iterate over
param_grid = {
    'vectorizer__max_features': [None],
    'model__max_depth': [None],
    'model__max_features': [5, 10],
    'model__n_estimators': [500, 1000]
}

print("Fitting model...")
#create and fit the grid and also time its fitting time
grid = GridSearchCV(pipeline, param_grid)
start_time = time.time()
grid.fit(tokenized_train, y_train)
end_time = time.time()

print("Done!")

#add the name of this run, the grid, the best params, and best score the lists to keep track
rf_names.append(prep_name + '_' + vectorizer_name + '_' + 'RandomForest')
rf_grids.append(grid)
rf_best_params.append(grid.best_params_)
rf_scores.append(grid.best_estimator_.score(tokenized_test, y_test))
rf_times.append(end_time-start_time)
```

```python
rf_results = pd.DataFrame({'model': rf_names,
                          'best_params': rf_best_params,
                          'train_times (s)': rf_times,
                          'best_score': rf_scores}).set_index('model').sort_values('best_score', ascending=False)

rf_results
```

```python
rf_best_params
```

```python
#now lets compare the performance of the best model for each classifier
nb_max_ind = np.argmax(nb_scores)
nb_preds = nb_grids[nb_max_ind].best_estimator_.predict(tokenized_test)
nb_f1, nb_precision, nb_recall = f1_score(y_test, nb_preds), precision_score(y_test, nb_preds), recall_score(y_test, nb_preds)
best_nb = [nb_names[nb_max_ind], nb_best_params[nb_max_ind], nb_times[nb_max_ind], nb_scores[nb_max_ind],
          nb_f1, nb_precision, nb_recall]


lgr_max_ind = np.argmax(lgr_scores)
lgr_preds = lgr_grids[lgr_max_ind].best_estimator_.predict(tokenized_test)
lgr_f1, lgr_precision, lgr_recall = f1_score(y_test, lgr_preds), precision_score(y_test, lgr_preds), recall_score(y_test, lgr_preds)
best_lgr = [lgr_names[lgr_max_ind], lgr_best_params[lgr_max_ind], lgr_times[lgr_max_ind], lgr_scores[lgr_max_ind],
           lgr_f1, lgr_precision, lgr_recall]

dtree_max_ind = np.argmax(dtree_scores)
dtree_preds = dtree_grids[dtree_max_ind].best_estimator_.predict(tokenized_test)
dtree_f1, dtree_precision, dtree_recall = f1_score(y_test, dtree_preds), precision_score(y_test, dtree_preds), recall_score(y_test, dtree_preds)
best_dtree = [dtree_names[dtree_max_ind], dtree_best_params[dtree_max_ind], dtree_times[dtree_max_ind], dtree_scores[dtree_max_ind],
             dtree_f1, dtree_precision, dtree_recall]

svm_max_ind = np.argmax(svm_scores)
svm_preds = svm_grids[nb_max_ind].best_estimator_.predict(tokenized_test)
svm_f1, svm_precision, svm_recall = f1_score(y_test, svm_preds), precision_score(y_test, svm_preds), recall_score(y_test, svm_preds)
best_svm = [svm_names[svm_max_ind], svm_best_params[svm_max_ind], svm_times[svm_max_ind], svm_scores[svm_max_ind],
           svm_f1, svm_precision, svm_recall]

rf_preds = rf_grids[0].best_estimator_.predict(tokenized_test)
rf_f1, rf_precision, rf_recall = f1_score(y_test, rf_preds), precision_score(y_test, rf_preds), recall_score(y_test, rf_preds)
best_rf = [rf_names[0], rf_best_params[0], rf_times[0], rf_scores[0], rf_f1, rf_precision, rf_recall]
```

```python
final_results = pd.DataFrame([best_nb, best_lgr, best_dtree, best_svm, best_rf], columns = ['model', 'best_params',
                                                                                            'train_time (s)', 
                                                                                            'accuracy_score', 'f1_score',
                                                                                            'precision_score',
                                                                                            'recall_score'
                                                                                           ]).sort_values('accuracy_score', ascending=False)

final_results
```
